## Part 8 — Written Summary

**Domain:** Agentic AI Course Content  
**User:** B.Tech students enrolled in the Agentic AI Hands-On Course by Dr. Kanthi Kiran Sirra  

**What the agent does:**  
The Course Assistant is a multi-turn agentic AI chatbot that answers questions about the Agentic AI course. It uses a LangGraph StateGraph with 8 nodes (memory, router, retrieval, skip, tool, answer, eval, save). It retrieves relevant course content from a ChromaDB vector database containing 12 focused documents. It uses three tools: DateTime (current date/time), Calculator (math expressions), and Web Search (live information). MemorySaver with thread_id enables multi-turn conversation memory with a 6-message sliding window. The eval_node self-reflects on faithfulness and retries answers scoring below 0.7.

**KB Size:** 12 documents, each 100–400 words, covering one topic each  
**Tools Used:** DateTime, Calculator, Web Search  

**RAGAS Baseline Scores:**

- Faithfulness: 0.71     (Hybrid system: RAG + Web)
